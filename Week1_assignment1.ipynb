### **Load the Dataset**

```python
# importing pandas module as pd (shortform) and then using read_csv function to read the csv from its file path
import pandas as pd

data = pd.read_csv('/content/sample_data/california_housing_train.csv')
data
```

### **Data Summary**

```python
# using describe function to find all stats for each column with numerical data
data.describe()
```

```python
# using loop i check for values and add it in the list, if its already present then it does nothing and moves ahead
list = []
for i in range (0,17000):
  n = data['total_bedrooms'][i]
  if n in list:
    pass
  else:
    list.append(n)

#print(list)
print("Distinct values of total_bedrooms: ",len(list))
```


### **Data Transformation**

```python
# using loop i calculate the reqd. value and add it to a list and then add a column to 'data' assigning it to the list
list2 = []
for i in range(0,17000):
  final = (data['total_bedrooms'][i]) / (data['total_rooms'][i])
  list2.append(final)

data['total_bedrooms_per_total_rooms'] = list2
data.head(2)
```

```python
# using describe to get a dataframe called summary which has all info and then extracting particular reqd. info from it
summary = data.describe()
print("Mean: ", summary['total_bedrooms_per_total_rooms']['mean'])
print("Standard Deviation: ", summary['total_bedrooms_per_total_rooms']['std'])
```

### **Data Filtering**


```python
# filtering those rows where median income is more than 5
dataset_1 = data[data['median_income']>5]
dataset_1
```

```python
# filtering based on conditions -- usage of & is necessary, 'and' won't work as these are not boolean values
dataset_2 = data[(data['total_rooms']>10000) & (data['median_house_value']<150000)]
dataset_2
```

### **Statistical Operations**

```python
# I create lists to store data for each column, then just create a loop to select each column calculate the stats
# and just add it to the respective list. A mask is created for each column and then applied on each column to
# obtain the array of reqd. values and convert it to list and add the values in this list to a master list - 'empty_list'
# finally i print all these lists and the empty_list as an array
mean_list = []
max_list = []
sum_list = []
empty_list = []

for i in range(0,4):
  mean = arr3_reshaped[:,i].mean()
  max = arr3_reshaped[:,i].max()
  sum = arr3_reshaped[:,i].sum()

  mask = arr3_reshaped[:,i]>mean

  l = (arr3_reshaped[:,i][mask]).tolist()
  for j in range(0, len(l)):
    empty_list.append(l[j])

  mean_list.append(mean)
  max_list.append(max)
  sum_list.append(sum)

print(mean_list, max_list, sum_list)
print(np.array(empty_list))
```

```python
#1 using infinite running loop to get the fibonacci nos (by adding the last 2 elements of the list)
# with a break condition that if the number to be appended > 100 then break the loop
list = [0,1]

while True:
  n = len(list)
  if list[n-1]+list[n-2] > 100:
    break
  list.append(list[n-1]+list[n-2])


print(list)
```
